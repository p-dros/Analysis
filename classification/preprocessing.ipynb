```python
import numpy as np

def check_normalization(data):
    # Oblicz średnią i odchylenie standardowe dla danych
    mean_value = np.mean(data)
    std_deviation = np.std(data)

    # Sprawdź zakres danych
    data_range = np.max(data) - np.min(data)

    # Wypisz wyniki
    print("Mean:", mean_value)
    print("Standard Deviation:", std_deviation)
    print("Data Range:", data_range)

    # Sprawdź, czy zakres danych jest bliski 1 (dla znormalizowanych danych)
    if np.isclose(data_range, 1.0):
        print("Data may be normalized.")
    else:
        print("Data may not be normalized.")

    # Sprawdź, czy średnia jest bliska zeru
    if np.isclose(mean_value, 0.0):
        print("Mean is close to zero.")
    else:
        print("Mean is not close to zero.")

    # Sprawdź, czy odchylenie standardowe jest bliskie jeden
    if np.isclose(std_deviation, 1.0):
        print("Standard Deviation is close to one.")
    else:
        print("Standard Deviation is not close to one.")

for column_name in df.columns:
    print(f"\nChecking normalization for column: {column_name}")
    check_normalization(df[column_name])
```

```python
correlation_matrix = df.corr()
```

```python
print("Most correlated pairs of attributes:")

correlation_pairs = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        attr1, attr2 = correlation_matrix.columns[i], correlation_matrix.columns[j]
        correlation_value = correlation_matrix.iloc[i, j]
        correlation_pairs.add(((attr1, attr2), correlation_value))

for pair in sorted(correlation_pairs, key=lambda x: abs(x[1]), reverse=True)[:10]:
    print(pair)


print("\n\nLeast correlated pairs of attributes")
for pair in sorted(correlation_pairs, key=lambda x: abs(x[1]), reverse=False)[:10]:
    print(pair)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_values = correlation_matrix.values.flatten()

plt.figure(figsize=(10, 6))
plt.hist(correlation_values, bins=50, alpha=0.7)
plt.xlabel('Correlation Value')
plt.ylabel('Frequency')
plt.title('Histogram of Correlation Values')
plt.show()
```

```python
num_pairs_over_05 = sum(abs(correlation_values) > 0.5)
print(f"Number of pairs with correlation value over 0.5: {num_pairs_over_05}")
```

```python
num_pairs_over_05 = sum(abs(correlation_values) > 0.75)
print(f"Number of pairs with correlation value over 0.75: {num_pairs_over_05}")
```

## Data preprocessing

### Removing correlated features

```python
threshold = 0.7
mask = (correlation_matrix.abs() > threshold) & (correlation_matrix != 1)

features_to_remove = set()
for feature in correlation_matrix.columns:
    correlated_features = correlation_matrix[mask[feature]].index.tolist()
    features_to_remove.update(correlated_features)

df_removed_correlated = df.drop(columns=features_to_remove)
print(f"Removed features: {features_to_remove}")
print(f"Amount of removed features: {len(features_to_remove)}")
print(list(df_removed_correlated))
```

### Checking for null values

```python
print("Amount of null values")
print(df_removed_correlated.isnull().sum().sum())
```

```python
df_removed_correlated.to_csv('data_proccessed.csv', index=False)
```

# Building models

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Bankrupt?'])
y = df['Bankrupt?']

rem_corr_df = df_removed_correlated
# .drop(columns=['Bankrupt?'])
rem_corr_train, rem_corr_test_all = train_test_split(df_removed_correlated, test_size = 0.3, random_state= 23)
rem_corr_test, rem_corr_valid = train_test_split(rem_corr_test_all, test_size=0.5, random_state=18)

X_train, X_test_all, y_train, y_test_all = train_test_split(X, y, test_size = 0.3, random_state= 23)
X_test, X_valid, y_test, y_valid = train_test_split(X_test_all, y_test_all, test_size=0.5, random_state=18)

print(X_train.shape[0],X_train.shape[0]/X.shape[0] )
print(X_test.shape[0], X_test.shape[0]/X.shape[0])
print(X_valid.shape[0], X_valid.shape[0]/X.shape[0])
print(f"Sum: {X_train.shape[0] + X_test.shape[0] + X_valid.shape[0]}")
print(f"Expected Sum: {X.shape[0]}")
```
